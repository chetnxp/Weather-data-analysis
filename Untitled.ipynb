```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MAHA_1994_2014.csv")
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("Columns in dataset:", df.columns.tolist())
```

```python
df.rename(columns={
    "max_temperature": "temperature_max",
    "min_temperature": "temperature_min",
    "mean_temperature": "temperature_mean",
    "precipitation_(mm)": "precipitation",
    "rain_(mm)": "rain",
    "snowfall_(cm)": "snowfall",
    "max_wind_speed": "wind_speed_max",
    "max_wind_gust": "wind_gust_max",
    "dominant_wind_direction": "wind_direction",
    "mean_humidity_(%)": "humidity_mean",
    "mean_soil_moisture_(%)": "soil_moisture_mean",
    "mean_solar_radiation_(w/m²)": "solar_radiation"
}, inplace=True)
```

```python
print("Columns in dataset:", df.columns.tolist())
```

```python
def compute_heatwave_index(temp_max, sealevel_pressure, humidity, solar_radiation, soil_moisture):
  
    norm_temp = (temp_max - 33.5) / (46.7 - 33.5)
    norm_temp = min(max(norm_temp, 0), 1)
    
    norm_pressure = (sealevel_pressure - 1000) / 20
    norm_pressure = min(max(norm_pressure, 0), 1)
    

    norm_humidity = (humidity - 70) / (100 - 70)
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_solar = (solar_radiation - 152.7) / (278.9 - 152.7)
    norm_solar = min(max(norm_solar, 0), 1)
    

    norm_soil = 1 - (soil_moisture / 0.5)
    norm_soil = min(max(norm_soil, 0), 1)
    
    weight_temp = 0.4
    weight_pressure = 0.2
    weight_humidity = 0.15
    weight_solar = 0.15
    weight_soil = 0.1

    raw_score = (weight_temp * norm_temp +
                 weight_pressure * norm_pressure +
                 weight_humidity * norm_humidity +
                 weight_solar * norm_solar +
                 weight_soil * norm_soil)
    
    
    heatwave_index = 1 + raw_score * 9
    return heatwave_index
```

```python
df['Heatwave_Index'] = df.apply(
    lambda row: compute_heatwave_index(
        row['temperature_max'],
        row['sealevel_pressure'],
        row['humidity_mean'],
        row['solar_radiation'],
        row['soil_moisture_mean']
    ),
    axis=1
)
```

```python
def compute_coldwave_index(temp_min, sealevel_pressure, snowfall, solar_radiation):
    """Computes a coldwave risk score based on relevant weather factors."""
    norm_temp = (21 - temp_min) / (21 - 6.1)  
    norm_temp = min(max(norm_temp, 0), 1)
    
    norm_pressure = (sealevel_pressure - 991) / (1021 - 991)
    norm_pressure = min(max(norm_pressure, 0), 1)
    
    norm_snow = snowfall / 42  
    norm_snow = min(max(norm_snow, 0), 1)
    
    norm_solar = 1 - ((solar_radiation - 100) / (278.9 - 100))  
    norm_solar = min(max(norm_solar, 0), 1)
    
    coldwave_index = 1 + (0.4 * norm_temp + 0.3 * norm_pressure + 0.2 * norm_snow + 0.1 * norm_solar) * 9
    return coldwave_index
```

```python
df["Coldwave_Index"] = df.apply(lambda row: compute_coldwave_index(
    row["temperature_min"], row["sealevel_pressure"], row["snowfall"], row["solar_radiation"]), axis=1)
```

```python
def compute_flood_index(precipitation, soil_moisture, humidity, surface_pressure):
    
    norm_precip = precipitation / 100  
    norm_precip = min(max(norm_precip, 0), 1)
    
    norm_soil = soil_moisture / 0.5  
    norm_soil = min(max(norm_soil, 0), 1)
    
    norm_humidity = (humidity - 60) / (100 - 60)  
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_pressure = 1 - ((surface_pressure - 950) / (1017 - 950))  
    norm_pressure = min(max(norm_pressure, 0), 1)
    
    flood_index = 1 + (0.5 * norm_precip + 0.2 * norm_soil + 0.2 * norm_humidity + 0.1 * norm_pressure) * 9
    return flood_index
```

```python
df["Flood_Index"] = df.apply(lambda row: compute_flood_index(
    row["precipitation"], row["soil_moisture_mean"], row["humidity_mean"], row["surface_pressure"]), axis=1)
```

```python
def compute_cloudburst_index(humidity, precipitation, temperature_mean, solar_radiation):
    
    norm_humidity = (humidity - 70) / (100 - 70)
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_precip = precipitation / 400  
    norm_precip = min(max(norm_precip, 0), 1)
    
    norm_temp = (temperature_mean - 23) / (39 - 23)  
    norm_temp = min(max(norm_temp, 0), 1)
    
    norm_solar = (solar_radiation - 150) / (278.9 - 150)
    norm_solar = min(max(norm_solar, 0), 1)
    
    cloudburst_index = 1 + (0.35 * norm_humidity + 0.35 * norm_precip + 0.2 * norm_temp + 0.1 * norm_solar) * 9
    return cloudburst_index
```

```python
df["Cloudburst_Index"] = df.apply(lambda row: compute_cloudburst_index(
    row["humidity_mean"], row["precipitation"], row["temperature_mean"], row["solar_radiation"]), axis=1)
```

```python
def compute_landslide_index(precipitation, soil_moisture):
   
    norm_precip = precipitation / 400
    norm_precip = min(max(norm_precip, 0), 1)
    
    norm_soil = soil_moisture / 0.5
    norm_soil = min(max(norm_soil, 0), 1)
    
    landslide_index = 1 + (0.7 * norm_precip + 0.3 * norm_soil) * 9
    return landslide_index
```

```python
df["Landslide_Index"] = df.apply(lambda row: compute_landslide_index(
    row["precipitation"], row["soil_moisture_mean"]), axis=1)
```

```python
def compute_drought_index(precipitation, temperature_max, humidity, soil_moisture):
    
    norm_precip = 1 - (precipitation / 50)  
    norm_precip = min(max(norm_precip, 0), 1)
    
    norm_temp = (temperature_max - 30) / (47 - 30)
    norm_temp = min(max(norm_temp, 0), 1)
    
    norm_humidity = 1 - (humidity / 70)  
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_soil = 1 - (soil_moisture / 0.5)
    norm_soil = min(max(norm_soil, 0), 1)
    
    drought_index = 1 + (0.4 * norm_precip + 0.3 * norm_temp + 0.2 * norm_humidity + 0.1 * norm_soil) * 9
    return drought_index
```

```python
df["Drought_Index"] = df.apply(lambda row: compute_drought_index(
    row["precipitation"], row["temperature_max"], row["humidity_mean"], row["soil_moisture_mean"]), axis=1)
```

```python
def compute_cyclone_index(wind_speed_max, wind_gust_max, humidity, sealevel_pressure):
    
    norm_wind = wind_speed_max / 50
    norm_wind = min(max(norm_wind, 0), 1)
    
    norm_gust = wind_gust_max / 120
    norm_gust = min(max(norm_gust, 0), 1)
    
    norm_humidity = (humidity - 60) / (100 - 60)
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_pressure = 1 - ((sealevel_pressure - 990) / (1015 - 990))
    norm_pressure = min(max(norm_pressure, 0), 1)
    
    cyclone_index = 1 + (0.4 * norm_wind + 0.3 * norm_gust + 0.2 * norm_humidity + 0.1 * norm_pressure) * 9
    return cyclone_index
```

```python
df["Cyclone_Index"] = df.apply(lambda row: compute_cyclone_index(
    row["wind_speed_max"], row["wind_gust_max"], row["humidity_mean"], row["sealevel_pressure"]), axis=1)
```

```python
def compute_thunderstorm_index(humidity, temperature_mean, wind_speed_max):
   
    norm_humidity = (humidity - 50) / (100 - 50)
    norm_humidity = min(max(norm_humidity, 0), 1)
    
    norm_temp = (temperature_mean - 20) / (39 - 20)
    norm_temp = min(max(norm_temp, 0), 1)
    
    norm_wind = wind_speed_max / 50  
    norm_wind = min(max(norm_wind, 0), 1)
    
    thunderstorm_index = 1 + (0.4 * norm_humidity + 0.3 * norm_temp + 0.3 * norm_wind) * 9
    return thunderstorm_index
```

```python
df["Thunderstorm_Index"] = df.apply(lambda row: compute_thunderstorm_index(
    row["humidity_mean"], row["temperature_mean"], row["wind_speed_max"]), axis=1)
```
